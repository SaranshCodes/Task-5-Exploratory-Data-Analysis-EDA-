```python
sns.pairplot(df[["Age","Fare","Pclass","Survived"]],hue="Survived")
```

### Pairwise Relationships Between Age, Fare, Pclass, and Survival
The pairplot shows pairwise relationships between selected numerical features (`Age`, `Fare`, `Pclass`) with survival status (`Survived`) as the hue: 

- **Age vs Survival:** Survivors (orange) are present across most age groups, but there is a visible cluster of survivors among younger passengers.
- **Fare vs Survival:** Higher fares are more associated with survivors, likely representing wealthier passengers in 1st class.
- **Pclass vs Survival:** Lower class numbers (1st class) show a higher proportion of survivors compared to 2nd and 3rd class.
- **Distributions:**
  - Age: Most passengers are between 20–40 years old, regardless of survival.
  - Fare: Highly skewed, with a few passengers paying significantly higher fares.
  - Pclass: Majority are in the 3rd class, with survival rates higher in the 1st class.

```python
corr=df.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.show()
```

### Correlation Heatmap
The heatmap below shows the correlation between numerical features in the dataset:

- **Survived** has a **negative correlation with Pclass (-0.34)**, indicating passengers in higher classes (lower Pclass number) had a higher survival rate.
- **Fare** is **positively correlated with Survived (0.26)** and **negatively correlated with Pclass (-0.55)**, meaning higher fares were generally paid by passengers in higher classes.
- **Age** shows a weak negative correlation with Survived (-0.077), suggesting younger passengers may have slightly higher survival rates, but the relationship is weak.
- **SibSp** and **Parch** are moderately correlated with each other (0.41), indicating that passengers with siblings/spouses onboard often had parents/children onboard too.
- Most other correlations are weak, suggesting minimal linear relationships between those features.

```python
sns.boxplot(x="Survived",y="Age",data=df)
plt.show()
```

### Age Distribution by Survival
The boxplot below compares the age distribution between passengers who survived and those who did not:

- **Median Age**: The median age of survivors and non-survivors is quite similar, around the late 20s.
- **Spread**: Both groups have a wide age range, from infants to elderly passengers.
- **Outliers**: Several older passengers (above 60) are present in both groups, showing that age alone was not a strict determinant of survival.
- **Slight Trend**: Survivors have a marginally higher proportion of younger passengers compared to non-survivors.
